```python
r = requests.get("https://api.sleeper.app/v1/draft/602012400345997312/picks")
data = r.json()
flat = json.dumps(data)
picks = pd.read_json(flat)
```

```python
picks_unrolled = pd.concat([picks.drop(['metadata'], axis=1), picks['metadata'].apply(pd.Series)], axis=1)
```
